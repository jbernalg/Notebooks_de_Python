# **Estructuras de control: Condicionales**

Un código, no deja de ser un conjunto de instrucciones que son ejecutadas una tras otra. Gracias a las estructuras de control, podemos cambiar el flujo de ejecución de un programa, haciendo que ciertos bloques de código se ejecuten sí y sólo si se dan unas condiciones paticulares.

## Condicionales

La estructura decisión lógica, mejor conocida como condicional , es utilizada para seleccionar la ruta que debe tomar la ejecución de instrucciones de un algoritmo.

### __1. Condicional Simple__

La estructura evalúa la condición y en tal caso:
<ul>
    <li>Si la condición es verdadera, entonces ejecuta la acción <b>True</b> (o acciones si son varias)</li>
    <li>Si la condición es falsa (<b>False</b>), entonces el algoritmo no se hace nada.</li>
</ul>


if_simple.png

Para la estructura _condicional simple_ en Python se utiliza la palabra reservada <b><i>if</i></b>, seguido va la condición que se evalúa, después de la condición se usan los <b><i>dos puntos</i></b> ($:$) y por último, para delimitar el código que está condicionado se usa la <b><i>indentación</i></b>.


if.png

```python
# ejemplo 
# definimos variables
a = 4
b = 2

# condicional simple 
if b != 0:
    print(a/b)
```

En el ejemplo anterior podemos ver cómo se puede usar un _if_ en Python. Con el operador _!=_ se comprueba que el número _b_ sea distinto de cero, y si lo es, se ejecuta el código que está indentado. Por tanto, el _if_ tiene dos partes:

- La condición que se tiene que cumplir para que el bloque de código se ejecute.
- El bloque de código que se ejecutará si se cumple la condición.

```python
print("Hello world!!!")

# definimos variable
x = 7
y = 5

# condicional simple
if y >= 15:
    print("VERDADERO !!!")


print("\nFin de programa")
```

También, se puede usar un _if_ dentro de otro _if_.

```python
num = 15
if num > 5:
    print('Es mayor que 5')
    if num <= 47:
        print('Está entre 5 y 47')
```

Se pueden usar los operadores lógicos (and, or y not) para concatenar condiciones.

```python
num = 12
if num > 5 and num <= 47:
    print('Está entre 5 y 47')
```

### __2. Condicional Doble__



<ul>
    <li>Si la condición a evaluar es verdadera, sigue el camino del <b>True</b> y se ejecutan las operaciones, en caso contrario sigue otro camino <b>False</b> y se ejecutan otras acciones.</li>
</ul>

if_doble.png



Para hablar de los _condicionales dobles_, se debe conocer la sentencia <b><i>else</i></b> de Python.

Una sentencia <b><i>else</i></b> le sigue a una sentencia <b><i>if</i></b> y contiene un código que es ejecutado cuando se evalúa la condición de la sentencia <b><i>if</i></b> y su resultado es <b><i>False</i></b>.
Al igual que la sentencia <b><i>if</i></b>, el código dentro del bloque debe ser indentado.



if_else.png

```python
# ejemplo

# definimos variable
Error = 25

# condicional doble
if Error <= 15:
    print("Error Aceptable...")
else:
    print("Se requieren ajustar los parámetros del modelo de Inteligencia Artificial")

print('Fin del algoritmo')
```

### __3. Condicional Compuestos o Anidados__

En la solución de problemas encontramos numerosos casos en los que luego de tomar una decisión y marcar el camino correspondiente a seguir, es necesario tomar otra decisión. Dicho proceso puede repetirse numerosas veces.

En aquellos problemas en donde un bloque condicional, en sus acciones incluye otro bloque condicional se dice que un bloque está anidado dentro del otro.

if_anidado.png

if_else_anidados.png

```python
# ejemplo 

# condicional anidado
if 1 + 1 == 2:
    if 2 * 2 == 8:
        print("Entró en el if")
    else:
        print("Entró en el else")
```

```python
# ejemplo

# definimos vaiable
num = 17

# condicional anidado
if num == 5:
    print('El número es 5')
else:
    if num == 11:
        print('El número es 11')
    else:
        if num == 7:
            print('El número es 7')
        else:
            print('El número no es 5, 7 u 11')
```

### __4. Condicional Múltiple__

Con frecuencia es necesario que existan más de dos elecciones posibles. Este problema se podría resolver por estructuras selectivas simples o dobles, anidadas o en cascada, pero si el número de alternativas es grande puede plantear serios problemas de escritura y de legibilidad.

if_multiple.png



<br><br>
Para trabajar las estructuras condicionales múltiples en Python, se utiliza la sentencia <b><i>elif</i></b>



if_elif_else.png

```python
num = int(input('Ingrese una de las opciones: '))
if num == 5:
    print('El número es 5')

elif num == 11:
    print('El número es 11')

elif num == 7:
    print('El número es 7')

else:
    print('El número no es 5, ni 7, ni 11')
```

### __5. Operador ternario__

Es una herramienta muy potente constituida por una clausula _if, else_ que se define en una sóla línea y puede ser usado por ejemplo dentro de un _print()_.

Existen tres partes en un operado ternario:

1. La condición a evaluar.
2. El código que se ejecuta si se cumple la condición.
3. El código que se ejecuta si no se cumple la condición.

> __[codigo si se cumple] if [condicion] else [codigo si no se cumple]__

```python
# ejemplo 

# definimos variable
x = 5

# operador ternario
print('Es 5' if x == 5 else 'No es 5')
```

Permite ahorrar algunas líneas de código, además de aumentar la rapidez con la que escribimos. Ejemplo, si tenemos una variable a la que queremos asigna un valor en función de una condición, se puede hacer como sigue:

```python
# ejemplo

# definimos variables
a = 10
b = 5

# operador ternario
c = a/b if b!=0 else -1
print(c)
```

Otro ejemplo es disminuir x en 1 si x es mayor a 0.

```python
# ejemplo 

# definimos variable
x = 5

# operador ternario
x -= 1 if x>0 else x
print(x)
```

### Ejercicio 10

Dado tres números enteros _a_, _b_ y _c_ y un objetivo target, encuentra cuál de los tres números es el más cercano al objetivo.

```python
# definimos los tres numeros
a, b, c = 7, 9, 7

# definimos el target
target = 10


# evaluamos las diferencias de a y b con el target c
diff_tc = target - c
diff_tb = target - b
diff_ta = target - a

# obtenemos el valor absoluto de ambas diferencias
if diff_ta < 0:
    diff_ta *= -1
elif diff_tb < 0:
    diff_tb *= -1
elif diff_tc < 0: 
    diff_tc *= -1

# buscamos la diferencia menor y devolvemos el numero correspondiente
if diff_ta < diff_tb and diff_ta < diff_tc:
    print(a)
elif diff_tb < diff_ta and diff_tb < diff_tc:
    print(b)
elif diff_tc < diff_ta and diff_tc < diff_tb:
    print(c)
elif diff_ta == diff_tb and diff_ta < diff_tc:
    print(a)
elif diff_tc == diff_tb and diff_tc < diff_ta:
    print(c)
elif diff_ta == diff_tc and diff_ta < diff_tb:
    print(c)
elif diff_tb == diff_ta and diff_tb == diff_tc:
    print(c)
```
