# **Estructuras de control: Condicionales**

Un codigo, no deja de ser un conjunto de instrucciones que son ejecutadas una tras ota. Gracias a las estructuras de control, podemos cambiar el flujo de ejecucion de un programa, haciendo que ciertos bloques de codigo se ejecuten si y solo si se dan unas condiciones paticulares

## Condicionales

La estructura decisión lógica, mejor conocida como condicional , es utilizada para seleccionar la ruta que debe tomar la ejecución de instrucciones de un algoritmo.

### __1. Condicional Simple__

La estructura evalúa la condición y en tal caso:
<ul>
    <li>Si la condición es verdadera, entonces ejecuta la acción <b>True</b> (o acciones si son varias)</li>
    <li>Si la condición es falsa (<b>False</b>), entonces el algoritmo no se hace nada.</li>
</ul>


if_simple.png

Para la estructura _condicional simple_ en Python se utiliza la palabra reservada <b><i>if</i></b>, seguido va la condición que se evalua, después de la condición se usan los <b><i>dos puntos</i></b> ($:$) y por último, para delimitar el código que está condicionado se usa la <b><i>indentación</i></b>.


if.png

```python
# ejemplo 1
# definimos variables
a = 4
b = 2

if b != 0:
    print(a/b)
```

En el ejemplo anterior podemos ver como se puede usar un _if_ en Python. Con el operador _!=_ se comprueba que el numero _b_ sea distinto de cero, y si lo es, se ejecuta el codigo que esta indentado. Por tanto el _if_ tiene dos partes:

- La condicion que se tiene que cumplir para que el bloque de codigo se ejecute.
- El bloque de codigo que se ejecutarar si se cumple la condicion.

```python
print("Hello world!!!")
x = 7
y = 5
if y >= 15:
    print("VERDADERO !!!")


print("\nFin de programa")
```

Tambien se puede usar un _if_ dentro de otro _if_.

```python
num = 15
if num > 5:
    print('Es mayor que 5')
    if num <= 47:
        print('Está entre 5 y 47')
```

Se pueden usar los operadores lógicos (and, or y not) para concatenar condiciones.

```python
num = 12
if num > 5 and num <= 47:
    print('Está entre 5 y 47')
```

### __2. Condicional Doble__



<ul>
    <li>Si la condición a evaluar es verdadera, sigue el camino del <b>True</b> y se ejecutan las operaciones, en caso contrario sigue otro camino <b>False</b> y se ejecutan otras acciones.</li>
</ul>

if_doble.png



Para hablar de los _condicionales dobles_, se debe conocer la sentencia <b><i>else</i></b> de Python.

Una sentencia <b><i>else</i></b> le sigue a una sentencia <b><i>if</i></b> y contiene un código que es ejecutado cuando se evalúa la condición de la sentencia <b><i>if</i></b> y su resultado es <b><i>False</i></b>.
Al igual que la sentencia <b><i>if</i></b>, el código dentro del bloque debe ser indentado.



if_else.png

```python
# ejemplo

# definimos variable
Error = 25

# condicional
if Error <= 15:
    print("Error Aceptable...")
else:
    print("Se requieren ajustar los parámetros del modelo de Inteligencia Artificial")

print('Fin del algoritmo')
```

### __3. Condicional Compuestos o Anidados__

En la solución de problemas encontramos numerosos casos en los que luego de tomar una decisión y marcar el camino correspondiente a seguir, es necesario tomar otra decisión. Dicho proceso puede repetirse numerosas veces.

En aquellos problemas en donde un bloque condicional, en sus acciones incluye otro bloque condicional se dice que un bloque está anidado dentro del otro.

if_anidado.png

if_else_anidados.png

```python
# ejemplo 1

if 1 + 1 == 2:
    if 2 * 2 == 8:
        print("Entró en el if")
    else:
        print("Entró en el else")
```

```python
# ejemplo 2

# definimos vaiable
num = 17

# condicional anidado
if num == 5:
    print('El número es 5')
else:
    if num == 11:
        print('El número es 11')
    else:
        if num == 7:
            print('El número es 7')
        else:
            print('El número no es 5, 7 u 11')
```

### __4. Condicional Múltiple__

Con frecuencia es necesario que existan más de dos elecciones posibles. Este problema se podría resolver por estructuras selectivas simples o dobles, anidadas o en cascada, pero si el número de alternativas es grande puede plantear serios problemas de escritura y de legibilidad.

if_multiple.png



<br><br>
Para trabajar las estructuras condicionales múltiples en Python, se utiliza la sentencia <b><i>elif</i></b>



if_elif_else.png

```python
num = int(input('Ingrese una de las opciones: '))
if num == 5:
    print('El número es 5')

elif num == 11:
    print('El número es 11')

elif num == 7:
    print('El número es 7')

else:
    print('El número no es 5, ni 7, ni 11')
```

### __5. Operador ternario__

Es una herramienta muy potente constituida por una clausula _if, else_ que se define en una sola linea y puede ser usado por ejemplo dentro de un _print()_.

Existen tres partes en un operado ternario:

1. La condicion a evaluar.
2. El codigo que se ejecuta si se cumple la condicion.
3. El codigo que se ejecuta si no se cumple la condicion.

> __[codigo si se cumple] if [condicion] else [codigo si no se cumple]__

```python
# ejemplo 1

# definimos variable
x = 5
print('Es 5' if x == 5 else 'No es 5')
```

Es muy util y permite ahorarse algunas lineas de codigo, ademas de aumentar la rapidez a la que escribimos. Ejemplo, si tenemos una variable a la que queremos asigna un valor en funcion de una condicion, se puede hacer como sigue:

```python
# ejemplo 2

# definimos variables
a = 10
b = 5

# operador ternario
c = a/b if b!=0 else -1
print(c)
```

Otro ejemplo es disminuir x en 1 si x es mayor a 0.

```python
# ejemplo 3

# definimos variable
x = 5

# operador ternario
x -= 1 if x>0 else x
print(x)
```

### Ejercicio 2

Compara tu edad con la de alguien mas. ¿Quién es mayor (tú o yo)?

```python
print('Quien es mayor, tu o yo?')

# solicita edad
age1 = int(input('Ingresa tu edad: '))
age2 = int(input('Ingresa la edad de la otra persona'))

# flujo de control
if age1 > 0 and age2 > 0:
    if age1 > age2:
        print(f'Tu eres {age1 - age2} años menor que yo!')
    if age1 < age2:
        print(f'Tu eres {age2 - age1} años mayor que yo!')
    if age1 == age2:
        print(f'tenemos la misma edad!')
else:
    print(f'Ingrese edades correctas')
```

### Ejecicio 5

Comprueba si la estación es Otoño, Invierno, Primavera o Verano a partir del mes dado. 
- Si el mes es Septiembre, Octubre o Noviembre, la estación es Otoño. 
- Si el mes es Diciembre, Enero o Febrero, la estación es Invierno. 
- Si el mes es Marzo, Abril o Mayo, la estación es Primavera. 
- Si el mes es Junio, Julio o Agosto, la estación es Verano.

```python
# solicitar mes
mes = input('Ingrese nombre del mes: ').lower()

# creamos listas de las estaciones y sus meses
otono = ['septiembre', 'obtubre', 'noviembre']
invierno = ['diciembre','enero', 'febrero']
primavera = ['marzo', 'abril', 'mayo']
verano = ['junio', 'julio', 'agosto']

# control de flujo
if mes in otono:
    print('Es Otoño!')
elif mes in invierno:
    print('Es invierno!')
elif mes in primavera:
    print('Es primavera!')
elif mes in verano:
    print('Es Verano!')
else:
    print('Ingrese un nombre de mes correcto!')
```

Construye un programa que, al recibir como dato el salario de un profesor de una universidad, calcule el incremento del salario de acuerdo con el siguiente criterio y escriba el nuevo salario del profesor.
