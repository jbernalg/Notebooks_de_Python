# **Estructuras de control: Condicionales**

Un cdigo, no deja de ser un conjunto de instrucciones que son ejecutadas una tras otra. Gracias a las estructuras de control, podemos cambiar el flujo de ejecucin de un programa, haciendo que ciertos bloques de cdigo se ejecuten s y slo si se dan unas condiciones paticulares.

## Condicionales

La estructura decisin lgica, mejor conocida como condicional , es utilizada para seleccionar la ruta que debe tomar la ejecucin de instrucciones de un algoritmo.

### __1. Condicional Simple__

La estructura evala la condicin y en tal caso:
<ul>
    <li>Si la condicin es verdadera, entonces ejecuta la accin <b>True</b> (o acciones si son varias)</li>
    <li>Si la condicin es falsa (<b>False</b>), entonces el algoritmo no se hace nada.</li>
</ul>


if_simple.png

Para la estructura _condicional simple_ en Python se utiliza la palabra reservada <b><i>if</i></b>, seguido va la condicin que se evala, despus de la condicin se usan los <b><i>dos puntos</i></b> ($:$) y por ltimo, para delimitar el cdigo que est condicionado se usa la <b><i>indentacin</i></b>.


if.png

```python
# ejemplo 
# definimos variables
a = 4
b = 2

# condicional simple 
if b != 0:
    print(a/b)
```

En el ejemplo anterior podemos ver cmo se puede usar un _if_ en Python. Con el operador _!=_ se comprueba que el nmero _b_ sea distinto de cero, y si lo es, se ejecuta el cdigo que est indentado. Por tanto, el _if_ tiene dos partes:

- La condicin que se tiene que cumplir para que el bloque de cdigo se ejecute.
- El bloque de cdigo que se ejecutar si se cumple la condicin.

```python
print("Hello world!!!")

# definimos variable
x = 7
y = 5

# condicional simple
if y >= 15:
    print("VERDADERO !!!")


print("\nFin de programa")
```

Tambin, se puede usar un _if_ dentro de otro _if_.

```python
num = 15
if num > 5:
    print('Es mayor que 5')
    if num <= 47:
        print('Est entre 5 y 47')
```

Se pueden usar los operadores lgicos (and, or y not) para concatenar condiciones.

```python
num = 12
if num > 5 and num <= 47:
    print('Est entre 5 y 47')
```

### __2. Condicional Doble__



<ul>
    <li>Si la condicin a evaluar es verdadera, sigue el camino del <b>True</b> y se ejecutan las operaciones, en caso contrario sigue otro camino <b>False</b> y se ejecutan otras acciones.</li>
</ul>

if_doble.png



Para hablar de los _condicionales dobles_, se debe conocer la sentencia <b><i>else</i></b> de Python.

Una sentencia <b><i>else</i></b> le sigue a una sentencia <b><i>if</i></b> y contiene un cdigo que es ejecutado cuando se evala la condicin de la sentencia <b><i>if</i></b> y su resultado es <b><i>False</i></b>.
Al igual que la sentencia <b><i>if</i></b>, el cdigo dentro del bloque debe ser indentado.



if_else.png

```python
# ejemplo

# definimos variable
Error = 25

# condicional doble
if Error <= 15:
    print("Error Aceptable...")
else:
    print("Se requieren ajustar los parmetros del modelo de Inteligencia Artificial")

print('Fin del algoritmo')
```

### __3. Condicional Compuestos o Anidados__

En la solucin de problemas encontramos numerosos casos en los que luego de tomar una decisin y marcar el camino correspondiente a seguir, es necesario tomar otra decisin. Dicho proceso puede repetirse numerosas veces.

En aquellos problemas en donde un bloque condicional, en sus acciones incluye otro bloque condicional se dice que un bloque est anidado dentro del otro.

if_anidado.png

if_else_anidados.png

```python
# ejemplo 

# condicional anidado
if 1 + 1 == 2:
    if 2 * 2 == 8:
        print("Entr en el if")
    else:
        print("Entr en el else")
```

```python
# ejemplo

# definimos vaiable
num = 17

# condicional anidado
if num == 5:
    print('El nmero es 5')
else:
    if num == 11:
        print('El nmero es 11')
    else:
        if num == 7:
            print('El nmero es 7')
        else:
            print('El nmero no es 5, 7 u 11')
```

Con frecuencia es necesario que existan ms de dos elecciones posibles. Este problema se podra resolver por estructuras selectivas simples o dobles, anidadas o en cascada, pero si el nmero de alternativas es grande puede plantear serios problemas de escritura y de legibilidad.

### __5. Operador ternario__

Es una herramienta muy potente constituida por una clausula _if, else_ que se define en una sla lnea y puede ser usado por ejemplo dentro de un _print()_.

Existen tres partes en un operado ternario:

1. La condicin a evaluar.
2. El cdigo que se ejecuta si se cumple la condicin.
3. El cdigo que se ejecuta si no se cumple la condicin.

> __[codigo si se cumple] if [condicion] else [codigo si no se cumple]__

```python
# ejemplo 

# definimos variable
x = 5

# operador ternario
print('Es 5' if x == 5 else 'No es 5')
```

Permite ahorrar algunas lneas de cdigo, adems de aumentar la rapidez con la que escribimos. Ejemplo, si tenemos una variable a la que queremos asigna un valor en funcin de una condicin, se puede hacer como sigue:

```python
# ejemplo

# definimos variables
a = 10
b = 5

# operador ternario
c = a/b if b!=0 else -1
print(c)
```

Otro ejemplo es disminuir x en 1 si x es mayor a 0.

```python
# ejemplo 

# definimos variable
x = 5

# operador ternario
x -= 1 if x>0 else x
print(x)
```
