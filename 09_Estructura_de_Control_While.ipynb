# **El ciclo $while$ en Python**

# Ciclos o bucles.

Los <b>ciclos</b> o <b>bucles</b> permiten ejecutar una o más instrucciones (secuenciales), cero o más veces, con base en el valor de verdad que arroje la evaluación de una expresión de tipo lógico. Esta expresión le permite al algoritmo tomar la decisión de repetir o dejar de ejecutar el grupo de instrucciones.

Python tiene implementado dos eestructuras de control cíclicas o de bucle, que son <b><i>while</i></b> y <b><i>for</i></b>

### Variable: $Contador$

Muchas veces en los procesos repetitivos es necesario hacer el conteo de sucesos o acciones internas del ciclo; este proceso de conteo se hace con una variable que se va incrementando cada vez que el ciclo se repite.

El <b><i>contador</i></b> es una variable que incrementa o disminuye su valor en forma constante cada que ocurre una acción o suceso.

La forma general del <b><i>contador</i></b> es:

$$CONTADOR\ =\ CONTADOR\ +\ <valor\ constante>$$

```python
cont = 0
print("Cont = ",cont)
cont = cont + 1 # cont = 0 + 1 = 1
print("Cont = ",cont)
cont = cont + 1 # cont = 1 + 1 = 2
print("Cont = ",cont)
cont = cont + 1 # cont = 2 + 1 = 3
print("Cont = ",cont)
cont = cont + 1 # cont = 3 + 1 = 4
print("Cont = ",cont)
```

### Variable: $Acumulador$

Un <b><i>acumulador</i></b> o totalizador es una variable cuya misión es almacenar cantidades variables resultantes de procesos sucesivos.

La diferencia con el contador radica en que el incremento o disminución es <b><i>variable</i></b> y NO constante.

La forma general del <b><i>acumulador</i></b> es:

$$ACUMULADOR\ =\ ACUMULADOR\ +\ <expresión>$$

```python
acum = 0
acum = acum + 4.5
acum = acum + 2.9
acum = acum + 3.6
acum = acum + 4.7
print("Acum = ",acum)
print("Promedio de notas = ",acum/4)
```

# Ciclo $while$

El ciclo _While_ permite ejecutar una seccion de codigo repetidas veces, de ahi su nombre. El codigo se ejecutara mientras una condicion determinada se cumpla. Cuando se deje de cumplir, se saldra del bucle y se continuara la ejecucion normal.

Llamemos __iteracion__ a una ejecucion completa del bloque de codigo. Analicemos el proceso que sigue el bucle _While_

Al ejecutarse esta instrucción, la condición es evaluada. Si la condición resulta verdadera ($True$), se ejecuta una vez la secuencia de instrucciones que forman el cuerpo del ciclo. Al finalizar la ejecución del cuerpo del ciclo se vuelve a evaluar la condición y si es verdadera la ejecución se repite, en caso contrario ($False$) se termina la ejecución del cuerpo del ciclo.

while.png



La palabra reservada en Python es <b><i>while</i></b>, luego la condición a evaluar, seguido se usan los dos puntos ( <b><i>:</i></b> ) y para delimitar el código que está incluido en el ciclo se usa la <b><i>indentación</i></b>.



while_text.png

```python
i = 1 # Inicializando el contador
while i <= 5: # Condición de parada
    print(i)
    i += 1 # CONTADOR -> Incremento del contador
           # i = i + 1

print("\nFin de programa ---- "+str(i))
```

Como se vio en el ejemplo anterior, el ciclo <b><i>while</i></b> se repite un número determinado de veces, porque la condición así lo establece. Pero, tambien es posible crear <b>ciclos</b> o <b>bubles infinitos</b>.

### Ciclos infinitos

Una forma de entrar en un bucle infinito es omitiendo el contador en las líneas indentadas el <b><i>while</i></b>

```python
i = 1 # Inicialización del contador
while i <= 5: # condición de parada
    print('Entró a un ciclo infinito - ' , i)

print("\nFin de programa")
```

```python
while 1==1: # SIEMPRE VA A SER TRUE
    print('Entró a un ciclo infinito')
```

Como se ve en el ejemplo anterior, no se usan variables en la estructura del <b><i>while</i></b>, todos son contantes, son elementos que no varían en el tiempo.

```python
k = 0 # inicialización del contador
while 3 != 2: # condición de parada -- SIEMPRE SERÁ TRUE
    print('Entró a un ciclo infinito - ' , k)
    k += 1 # CONTADOR
```

En la condición del ciclo <b><i>while</i></b> se pueden usar otros tipos datos o variables como <b><i>cadenas</i></b>  y <b><i>booleanos</i></b>

```python
# Usando datos tipos cadena, también llamados alfanuméricos
m = 0
while 'a' != 'b': # Esto siempre va a ser un True
    print('Entró a un ciclo infinito - ' , m)
    m += 1

print('\nFin de programa.')
```

```python
# Usando datos tipos lódicos o booleanos
m = 0
while True:
    print('Entró a un ciclo infinito - ' , m)
    m += 1

print('\nFin de programa.')
```

# Sentencia $break$ y $continue$ en el Ciclo $while$</h4>


* Para finalizar un bucle while prematuramente, se puede utilizar la sentencia <b><i>break</i></b>.

* Para volver a la primera línea del ciclo debido a una condición dada, se usa el <b><i>continue</i></b>.

```python
i = 0 # inicialización del contador
while True:
    i = i + 1 # contador - incremento constante de la variable

    if i == 2:
        print ('Paso 2')
        continue
    if i == 5:
        print ('Breaking...')
        break
    print(i)
print('Fin de programa')
```

# Ciclo $while$ en una linea

Es posible tener un _while_ es una sola linea, algo muy util si el bloque que queremos ejecutar es corto. En el caso de tener mas de una sentencia, las debemos separar con ;

```python
# contador
x = 5

# ciclo while
while x > 0: x-=1; print(x)
```

# $Else$ y $while$

Python admite el uso de la clausula _else_ al final de _while_. La seccion de codigo que se encuentra dentro de _else_, se ejcutara cuando el bucle termine, pero solo si lo hace por razones naturales. Es decir, si el bucle termina porque la condicion se deja de cumplir y no porque se ha hecho uso del _break_

```python
# contador
x = 5

# ciclo while
while x > 0:
    x -= 1
    print(x)
else:
    print('El bucle ha terminado!')
```

En el siguiente codigo el bloque _else_ no se ejecuta debido a que el ciclo termina por _break_.

```python
# contador
x = 5

# ciclo while
while x > 0:
    x -= 1
    print(x)
    # condicion que rompe el ciclo
    if x == 0:
        break
else:
    print('El bucle ha terminado!')
```

### Exercise 1
Diseñe un algoritmo en Python que calcule el promedio de los errores de clasificación que comete el algoritmo de Inteligencia Arficial en los diferentes ciclos de entrenamiento. El algoritmo deberá pedir al usuario ingresar el número de errores que va a promediar y este será el número de veces que se repita el ciclo while pidiendo la magnitud del error y haciendo el cálculo del promedio.

```python
n_errors = int(input('Ingrese numero de errores: '))

# inicializar bandera
i = 1

# acumulador
sum_err = 0

# ciclo while
while i <= n_errors:
    e = float(input('Ingrese el error medido: '))
    sum_err = sum_err + e
    i = i + 1
    
# calculo promedio del error
prom = sum_err/n_errors

# mostrar promedio
print(f'El promedio de los errores es: {prom}')
```

### Exercise 2
Escribe un programa que encuentre los números que son divisibles por 7 y múltiplos de 5, entre 1500 y 2700 (ambos incluidos).

```python
# contador
count = 1500

# lista para almacenar numeros
nums = []


# ciclo
while count <= 2700:

    if count%7 == 0 and count%5 == 0:
        nums.append(count)
        
    count += 1
    

# mostrar resultados
print('Numeros divisibles por 7 y multiplo de 5: ', nums)
```

### Exercise 3
Escriba un programa Python para adivinar un número entre 1 y 9. Puntos importantes:

- Se le solicita al usuario que ingrese una suposición. 

- Si la suposición es incorrecta, el mensaje aparece nuevamente hasta que la suposición sea correcta. 

- Si la suposición es correcta, el usuario recibirá un mensaje que dice "¡Bien adivinado!" y el programa se cerrará.

- El usuario tendra un maximo de 3 oportunidades.

- No se admitira numero repetidos.

```python
# numero seleccionado
num_Adivinar = 7
# oportunidades
opp = 3
# lista para almacenar numeros dados por el usuario
num_dados = []
# numero supuesto por el ususario
num_supuesto = int(input('Ingrese un numero entre 1 y 9: '))

while True:

    if num_Adivinar == num_supuesto:
        print('Bien. Lo has adivinado!')
        break
    else:
        if num_supuesto in num_dados:
            print(f'El numero {num_supuesto} ya fue ingresado!')
        else:
            num_dados.append(num_supuesto)
            opp = opp - 1
    
    if opp != 0:
        print(f'Te quedan {opp} oportunidades!')
    else:
        print('Has perdido!')
        break
    
    num_supuesto = int(input('Ingrese nuevamente un numero entre 1 y 9: '))
    
```

### Exercise 10

### Exercise 12
Escriba un programa en Python que acepte una cadena y calcule la cantidad de dígitos y letras.

```python
sentence = input()
print(sentence)

band = 0
letters, digit = 0, 0

while band < len(sentence):

    if sentence[band].isalpha():
        letters += 1
    elif sentence[band].isdigit():
        digit += 1

    band += 1

print(f'Cantidad de letras: {letters}')
print(f'Cantidad de digitos: {digit}')
```
