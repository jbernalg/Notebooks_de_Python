# **El ciclo $while$ en Python**

# Ciclos o bucles.

Los <b>ciclos</b> o <b>bucles</b> permiten ejecutar una o ms instrucciones (secuenciales), cero o ms veces, con base en el valor de verdad que arroje la evaluacin de una expresin de tipo lgico. Esta expresin le permite al algoritmo tomar la decisin de repetir o dejar de ejecutar el grupo de instrucciones.

Python tiene implementado dos eestructuras de control cclicas o de bucle, que son <b><i>while</i></b> y <b><i>for</i></b>

### Variable: $Contador$

Muchas veces en los procesos repetitivos es necesario hacer el conteo de sucesos o acciones internas del ciclo; este proceso de conteo se hace con una variable que se va incrementando cada vez que el ciclo se repite.

El <b><i>contador</i></b> es una variable que incrementa o disminuye su valor en forma constante cada que ocurre una accin o suceso.

La forma general del <b><i>contador</i></b> es:

$$CONTADOR\ =\ CONTADOR\ +\ <valor\ constante>$$

```python
cont = 0
print("Cont = ",cont)
cont = cont + 1 # cont = 0 + 1 = 1
print("Cont = ",cont)
cont = cont + 1 # cont = 1 + 1 = 2
print("Cont = ",cont)
cont = cont + 1 # cont = 2 + 1 = 3
print("Cont = ",cont)
cont = cont + 1 # cont = 3 + 1 = 4
print("Cont = ",cont)
```

### Variable: $Acumulador$

Un <b><i>acumulador</i></b> o totalizador es una variable cuya misin es almacenar cantidades variables resultantes de procesos sucesivos.

La diferencia con el contador radica en que el incremento o disminucin es <b><i>variable</i></b> y NO constante.

La forma general del <b><i>acumulador</i></b> es:

$$ACUMULADOR\ =\ ACUMULADOR\ +\ <expresin>$$

```python
acum = 0
acum = acum + 4.5
acum = acum + 2.9
acum = acum + 3.6
acum = acum + 4.7
print("Acum = ",acum)
print("Promedio de notas = ",acum/4)
```

# Ciclo $while$

El ciclo _While_ permite ejecutar una seccin de cdigo repetidas veces, de ah su nombre. El cdigo se ejecutar mientras una condicin determinada se cumpla. Cuando se deje de cumplir, se saldr del bucle y se continuar la ejecucin normal.

Llammos __iteracin__ a una ejecucin completa del bloque de cdigo. Analicmos el proceso que sigue el bucle _While_.

Al ejecutarse esta instruccin, la condicin es evaluada. Si la condicin resulta verdadera ($True$), se ejecuta una vez la secuencia de instrucciones que forman el cuerpo del ciclo. Al finalizar la ejecucin del cuerpo del ciclo se vuelve a evaluar la condicin y si es verdadera la ejecucin se repite, en caso contrario ($False$) se termina la ejecucin del cuerpo del ciclo.

while.png



La palabra reservada en Python es <b><i>while</i></b>, luego la condicin a evaluar, seguido se usan los dos puntos ( <b><i>:</i></b> ) y para delimitar el cdigo que est incluido en el ciclo se usa la <b><i>indentacin</i></b>.



while_text.png

```python
i = 1 # Inicializando el contador
while i <= 5: # Condicin de parada
    print(i)
    i += 1 # CONTADOR -> Incremento del contador
           # i = i + 1

print("\nFin de programa ---- "+str(i))
```

Como se vi en el ejemplo anterior, el ciclo <b><i>while</i></b> se repite un nmero determinado de veces, porque la condicin as lo establece. Pero, tambin es posible crear <b>ciclos</b> o <b>bubles infinitos</b>.

### Ciclos infinitos

Una forma de entrar en un bucle infinito es omitiendo el contador en las lneas indentadas el <b><i>while</i></b>

```python
i = 1 # Inicializacin del contador
while i <= 5: # condicin de parada
    print('Entr a un ciclo infinito - ' , i)

print("\nFin de programa")
```

```python
while 1==1: # SIEMPRE VA A SER TRUE
    print('Entr a un ciclo infinito')
```

Como se ve en el ejemplo anterior, no se usan variables en la estructura del <b><i>while</i></b>, todos son contantes, son elementos que no varan en el tiempo.

```python
k = 0 # inicializacin del contador
while 3 != 2: # condicin de parada -- SIEMPRE SER TRUE
    print('Entr a un ciclo infinito - ' , k)
    k += 1 # CONTADOR
```

En la condicin del ciclo <b><i>while</i></b> se pueden usar otros tipos datos o variables como <b><i>cadenas</i></b>  y <b><i>booleanos</i></b>

```python
# Usando datos tipos cadena, tambin llamados alfanumricos
m = 0
while 'a' != 'b': # Esto siempre va a ser un True
    print('Entr a un ciclo infinito - ' , m)
    m += 1

print('\nFin de programa.')
```

```python
# Usando datos tipos ldicos o booleanos
m = 0
while True:
    print('Entr a un ciclo infinito - ' , m)
    m += 1

print('\nFin de programa.')
```

# Sentencia $break$ y $continue$ en el Ciclo $while$</h4>


* Para finalizar un bucle while prematuramente, se puede utilizar la sentencia <b><i>break</i></b>.

* Para volver a la primera lnea del ciclo debido a una condicin dada, se usa el <b><i>continue</i></b>.

```python
i = 0 # inicializacin del contador
while True:
    i = i + 1 # contador - incremento constante de la variable

    if i == 2:
        print ('Paso 2')
        continue
    if i == 5:
        print ('Breaking...')
        break
    print(i)
print('Fin de programa')
```

# Ciclo $while$ en una lnea

Es posible tener un _while_ es una sla lnea, algo muy til si el bloque que queremos ejecutar es corto. En el caso de tener ms de una sentencia, las debemos separar con ';'

```python
# contador
x = 5

# ciclo while
while x > 0: x-=1; print(x)
```

# $Else$ y $while$

Python admite el uso de la clasula _else_ al final de _while_. La seccin de cdigo que se encuentra dentro de _else_, se ejecutar cuando el bucle termine, pero slo si lo hace por razones naturales. Es decir, si el bucle termina porque la condicin se deja de cumplir y no porque se ha hecho uso del _break_.

```python
# contador
x = 5

# ciclo while
while x > 0:
    x -= 1
    print(x)
else:
    print('El bucle ha terminado!')
```

En el siguiente cdigo el bloque _else_ no se ejecuta debido a que el ciclo termina por _break_.

```python
# contador
x = 5

# ciclo while
while x > 0:
    x -= 1
    print(x)
    # condicion que rompe el ciclo
    if x == 0:
        break
else:
    print('El bucle ha terminado!')
```

# Bucles Anidados

Podemos agregar un _while_ dentro de otro, para realizar operaciones ms complejas. Es especialmente til para generar permutaciones de numeros.

En el siguiente ejemplo vamos a generar todas las combinaciones de dos numeros hasta 2.

```python
# permutacion de 0-0 a 2-2
i, j = 0, 0

while i < 3:
    while j < 3:
        print(i, j)
        j += 1
    i += 1
    j = 0
```

Vamos a generar combinaciones de 3 elementos con los numeros 0, 1 y 2.

```python
# permutacion de 0-0-0 a 2-2-2

i, j, k = 0, 0, 0

while i < 3:
    while j < 3:
        while k < 3:
            print(i,j,k)
            k += 1
            j += 1
        k = 0
    i += 1
    j = 0
```

### Ejercicio 1
Disee un algoritmo en Python que calcule el promedio de los errores de clasificacin que comete el algoritmo de Inteligencia Arficial en los diferentes ciclos de entrenamiento. El algoritmo deber pedir al usuario ingresar el nmero de errores que va a promediar y este ser el nmero de veces que se repita el ciclo while pidiendo la magnitud del error y haciendo el clculo del promedio.

```python
n_errors = int(input('Ingrese numero de errores: '))

# inicializar bandera
i = 1

# acumulador
sum_err = 0

# ciclo while
while i <= n_errors:
    e = float(input('Ingrese el error medido: '))
    sum_err = sum_err + e
    i = i + 1
    
# calculo promedio del error
prom = sum_err/n_errors

# mostrar promedio
print(f'El promedio de los errores es: {prom}')
```

### Ejercicio 2
Escribe un programa que encuentre los nmeros que son divisibles por 7 y mltiplos de 5, entre 1500 y 2700 (ambos incluidos).

```python
# contador
count = 1500

# lista para almacenar numeros
nums = []


# ciclo
while count <= 2700:

    if count%7 == 0 and count%5 == 0:
        nums.append(count)
        
    count += 1
    

# mostrar resultados
print('Numeros divisibles por 7 y multiplo de 5: ', nums)
```

Escriba un programa Python para adivinar un nmero entre 1 y 9. Puntos importantes:

- Se le solicita al usuario que ingrese una suposicin. 

- Si la suposicin es incorrecta, el mensaje aparece nuevamente hasta que la suposicin sea correcta. 

- Si la suposicin es correcta, el usuario recibir un mensaje que dice "Bien adivinado!" y el programa se cerrar.

### Ejercicio 4
Escriba un programa para comprobar la validez de las contrasenas ingresadas por los usuarios.

Valicacion:

- Al menos 1 letra entre [az] y 1 letra entre [AZ]

- Al menos 1 numero entre [0-9].

- Al menos 1 caracter de [$#@].

- Longitud minima 6 caracteres.

- Longitud maxima 16 caracteres.

```python
# importamos la libreria re para expresiones regulares
import re

# solicitamos la contrasena
password = input('Ingrese la contrasena: ')

# creamos la bandera del bucle
x = True

# inicia el bucle mientras x sea True
while x:

    # verificamos las condiciones de una contrasena valida

    # longitud de la contrsena
    if (len(password) < 6 or len(password) > 12):
        break
    # contrasena con al menos 1 letra minuscula
    elif not re.search("[a-z]", password):
        break
    # contrasena con al menos 1 letra mayuscula
    elif not re.search("[A-Z]", password):
        break
    # contrasena con al menos 1 digito
    elif not re.search("[0-9]", password):
        break
    # contrasena con al menos un caracter especial $#@
    elif not re.search("[$#@]", password):
        break
    # contrasena sin espacios en blanco
    elif re.search("\s", password):
        break
    else:
        # contrasena valida
        print('Contrasea Valida')
        # cambiamos valor de la bandera para salir del bucle
        x = False
        break

# si la bandera sigue siendo True, la contrasena no es valida
if x:
    print('Contrasea no valida')
```

### Ejercicio 5
Pide al usuario ingresar un nmero y verifica si es primo o no. El programa debe continuar hasta que el usuario decida salir.

```python
# definimos la variable primo como True
es_primo = True

# definimos el ciclo principal
while True:
    
    # solicita al usuario un numero entero
    num = int(input('Ingrese un numero entero para verificar si es primo: '))

    # verifica si el usuario desea salir
    if num == -1:
        print('Fin del programa!')
        break

    # determinar si el numero es primo
    if num > 1:
        # definimos a divisor como contador
        divisor = 2
        # verificamos el divisor hasta la raiz cuadrada del numero para optimizar
        while divisor <= int(num**0.5):
            if num % divisor == 0:
                es_primo = False
                break
            divisor += 1

        # mostramos resultados
        if es_primo:
            print(f'{num} es un numero primo')
        else:
            print(f'{num} no es un numero primo')    

```

### Ejercicio 6
Solicita al usuario ingresar un nmero entero y calcula la suma de sus dgitos. Repite el proceso hasta que el usuario decida detenerse.

```python
# iniciamos el ciclo while
while True:

    # solicitamos el numero entero
    num = input('Ingrese un numero entero (o escribe "salir" para terminar el programa): ')

    # verifica si el usuario ddesea salir
    if num.lower() == 'salir':
        print('Fin del programa!')
        break

    # validamos que el numero sea un entero
    if num.isdigit() or (num[0] == '-' and num[1:].isdigit()):
        num = int(num)
        # sumamos los digitos
        suma_digitos = sum(int(digito) for digito in str(abs(num)))
        # mostrar resultado
        print(f'La suma de los digitos de {num} es {suma_digitos}')
    else:
        print('Por favor, ingresa un numero valido!')
```

### Ejercicio 7
Crea un programa que permita al usuario ingresar un saldo inicial. Luego, ofrece opciones para depositar, retirar dinero o consultar el saldo. Usa un bucle para repetir estas operaciones hasta que el usuario elija salir.

```python
# inicializamos el saldo inicial
saldo = float(input('Ingrese el saldo inicial: '))

# iniciamos el bucle while
while True:
    # mostramos las opciones
    print('\n--- Menu de opciones ---')
    print('1. Consultar saldo')
    print('2. Depositar dinero')
    print('3. Retirar dinero')
    print('4. Salir')

    # Solicitar eleccion del usuario
    opcion = input('Seleccione una opcion (1-4): ')

    # consultar saldo
    if opcion == '1':
        print(f'Su saldo actual es: ${saldo:.2f}')

    # depositar dinero
    elif opcion == '2':
        deposito = float(input('Ingrese la cantidad a depositar: '))
        if deposito > 0:
            saldo += deposito
            print(f'Has depositado ${deposito:.2f}. Su nuevo saldo es: ${saldo:.2f}')

    # retirar dinero
    elif opcion == '3':
        retiro = float(input('Ingrese la cantidad a retirar: '))
        if retiro > saldo:
            print('Fondos insuficientes!')
        elif retiro > 0:
            saldo -= retiro
            print(f'Has retirado ${retiro:.2f}. Su nuevo saldo es: ${saldo:.2f}')
        else:
            print('El monto a retirar debe ser positivo!')

    # salir del programa
    elif opcion == '4':
        print('Hasta Pronto!')
        break

    else:
        # opciones no validas
        print('Opcion no valida! Intente de nuevo!')
```

### Ejercicio 8
Implementa un programa que genere contraseas aleatorias de una longitud definida por el usuario. Permite que el usuario genere mltiples contraseas hasta que decida salir.

```python
# importamos las librerias random para manejar caracteres aleatorios
import random

# importamos string para manejar conjuntos predefinidos de caracteres
import string

print('===== Generador de contraseas =====')

# iniciamos el bucle
while True:
    # solicita la longitud de la contrasena
    longitud = input('Ingrese la longitud de la contrasena (o "salir" para terminar): ')

    # verificamos si el usuario desea salir
    if longitud.lower() == 'salir':
        print('Fin del programa!')
        break

    # verificamos que la longitud ingresada sea correcta
    if not longitud.isdigit() or int(longitud) <= 0:
        print('Ingrese un valor de longitud valido!')
        continue

    longitud = int(longitud)

    # generamos la contrasena aleatoria
    caracteres = string.ascii_letters + string.digits + string.punctuation
    contrasena = ''.join(random.choice(caracteres) for _ in range(longitud))

    # mostrar contrasena generada
    print(f'Contrasena generada: {contrasena}')
```

### Ejercicio 12
Escriba un programa en Python que acepte una cadena y calcule la cantidad de dgitos y letras.

```python
sentence = input()
print(sentence)

band = 0
letters, digit = 0, 0

while band < len(sentence):

    if sentence[band].isalpha():
        letters += 1
    elif sentence[band].isdigit():
        digit += 1

    band += 1

print(f'Cantidad de letras: {letters}')
print(f'Cantidad de digitos: {digit}')
```
