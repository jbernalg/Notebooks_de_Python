## 2. Instruccin de entrada en Python - $input()$

La funcin __input()__ permite obtener texto escrito por teclado. Al llegar a la funcin, el programa (algoritmo) se detiene esperando que se escriba algo y se presione la tecla _Enter_, como muestra el siguiente ejemplo:

```python
print("Cul es su nombre?")
nombre = input()
print("Me alegro de conocerlo/a " + nombre)
```

En el ejemplo anterior, el usuario escribe la respuesta de su nombre en una lnea diferente a la que aparece la pregunta. Pero, se puede hacer el algoritmo para que todo quede en el mismo rengln. Miremos el siguiente ejemplo:

```python
nombre = input("Cul es su nombre?: ")
print("Me alegro de conocerlo/a " + nombre)
```

__IMPORTANTE__

De forma predeterminada, la funcin __input()__ captura la entrada y la convierte en una cadena. Si el dato que se ingresa numrico, esta funcin los convierte en cadena. Si se intenta hacer operaciones aritmticas, puede producir un error.

```python
a = input("Ingrese un nmero: ")
b = input("Ingrese otro nmero: ")
print("El resultado de la suma es: " + str(a + b))
```

```python
# verifiquemos el tipo de dato del numero b con type
type(b)
```

En el ejemplo anterior no se presenta error porque el operado cruz ( $+$ ) se comparte para la $suma$ en las operaciones aritmticas y $concatenacin$ en las operaciones alfanumricas. Pero, Qu tal si fuera el producto o la divisin o algn signo que no se comparta?

```python
a = input("Ingrese un nmero: ")
b = input("Ingrese otro nmero: ")
print("El resultado de la multiplicacin es: " + str(a * b))
```

> El error anterior se produce cuando se usan los operadores aritmticos: $ -\ ,\ *,\ /,\,\ **\ , \%,\ //$

Si lo que se requiere es usar las entradas en operaciones aritmticas, se puede hacer la conversin en el momento de hacer la captura del dato.

```python
# Usando la funcin int()
a = int(input("Ingrese un nmero: "))
b = int(input("Ingrese otro nmero: "))
print("El resultado de la suma es: " + str(a + b))
```

```python
# Usando la funcin float()
a = float(input("Ingrese un nmero: "))
b = float(input("Ingrese otro nmero: "))
print("El resultado de la suma es: " + str(a + b))
```

### 3.2. Funcin __format()__

La forma general de este mtodo es: $str.format()$

```python
# Uso 1
print("Saludo a {} y {} - {}".format('Bruce Wayne','Clark Kent' , 987))
```

```python
variable_format = "Saludo a los {} de {} - {}".format('Bruce Wayne','Clark Kent' , 987)
print(variable_format)
```

Las llaves $\{\ \}$ y caracteres dentro de las mismas (llamados campos de formato) son reemplazadas con los valores pasados en el mtodo $str.format()$. Un nmero en las llaves se refiere a la posicin del valor pasado en el mtodo $str.format()$.

```python
# Uso 2
name = input('Nombre: ')
lastname = input('Apellidos: ')
age = int(input('Edad: '))
print('Hola {0} {1}, tienes {2} aos de edad'.format(name,lastname,age))
```

Los ndices de las llaves $\{\ \}$ se pueden usar tantas veces como sea necesario y en el orden deseado, siempre y cuando exista en los valores pasados por el mtodo $str.format()$

```python
print('{0}{1}{0}'.format('abra', 'cad') )
```

Tambin se puede usar con variables, como se muestra a continuacin

```python
name = "Peter"
lastname = "Quill"
edad = 36
print('Hola {name} {lastname}, tienes {edad} aos de edad'.format(name=name,lastname=lastname,edad=edad))
```
