## Eliminar Elementos

Al momento de eliminar un dato de la lista, esta se reacomodar.

Con la reacomodacin, el tamao de la lista se ver reducido en uno y los elementos ubicados a la derecha del dato eliminado cambiarn su subndice. Para remover un dato de una lista tenemos varias alternativas.

```python
frutas = ['manzana', 'durazno', 'pera', 'pia']
frutas
```

```python
# eliminar el elemento 'pera'
del frutas[2]
frutas
```

```python
# eliminar todos los elementos excepto el primero
del frutas[1:]
frutas
```

## Secciones de lista

Podemos crear "secciones" o sublistas a partir de una lista existente.

Una sublista es una lista que se crea a partir de otra lista. Al crearse, siempre tendr elementos en comn con la lista "padre".

Su sintaxis es similar a la de acceder a un dato.

```python
frutas = ['manzana', 'durazno', 'pera', 'pia', 'ciruela', 'kiwi']
frutas
```

```python
# acceder a los dos elementos de la mitad de la lista
frutas[2:4]
```

Una sublista tiene una posicin inicial y una posicin final. As que, en lugar de indicar un nico valor numrico en los corchetes, debemos indicarlo de la siguiente manera:

1.   subndice del primer elemento de la sublista
2.   operador dos puntos (:)
3.   subndice del _ltimo-1_ elemento de la sublista

Esto quiere decir que el primer valor es **inclusivo** y el ltimo valor es **exclusivo**.

En la celda anterior, la sintaxis fue la siguiente:

*   2
*   :
*   4

Si restamos el segundo valor con el primer valor `(4 - 2)`, obtenemos la cantidad de elementos de la sublista (si tienen el mismo signo). En este caso, fueron dos: `'pera'` y `'pia'`.




```python
# acceder a los 4 elementos desde la posicion 1
frutas[1:5]
```

Si omitimos el primer subindice, toma por defecto la posicin 0.

```python
frutas[:2]
```

Si omitimos el ltimo subndice, tomar como valor por defecto la cantidad de elementos de la lista padre.

```python
frutas[3:]
```

Incluso si omitimos ambos subndices, tendrn los mismos valores por defecto de la lista padre.

```python
frutas[:]
```

Si pasamos un valor de subndice mayor que la cantidad de elementos de la lista, devolver slo los elementos de la lista.

```python
frutas[:20]
```

```python
frutas[2:40]
```

Podemos utilizar valores negativo en la sintaxis para acceder a los elementos de forma inversa.

```python
# ultimo elemento de la lista
frutas[-1:]
```

```python
# acceder a elemento de forma inversa
frutas[-5:-2]
```

Si a la sintaxis $[:]$ aadimos otro operador dos puntos (:) y un valor numrico, se determinar la secuencia que utilizar la sublista para aadir los elementos. Por ejemplo, $[::2]$ tendr en cuenta todos los elementos de la lista y los aadir a la sublista de dos en dos.

```python
frutas = ['manzana', 'durazno', 'pera', 'pia', 'ciruela', 'kiwi']
frutas
```

```python
# mostrar los elementos de cada dos posiciones
frutas[::2]
```

Cualquier secuencia puede utilizar la sintaxis de secciones, como el *string*

```python
'Mesada'[:4]
```

## Compresin de Listas

Ofrecen una manera concisa de crear listas. Crea nuevas listas donde cada elemento es el resultado de algunas operaciones aplicada a cada mienbro de otra secuencia o iterable.

```python
# aplicar el valor absoluto sobre todos los valores del vector
[abs(x) for x in vec]
```

### $filter()$

Filtra la lista segn una funcin que devuelve verdadero o falso. Debemos guardar el resultado de filter en una lista para poder visualizarlo.

```python
# funcion que determina si un numero es par
def es_par(n):
    return n % 2 == 0
```

```python
# lista de numeros
numeros = [1, 7, 3, 22, 5, 11, 5, 8, 12, 0]
# aplicar filter, pasar los valores a una lista y mostrarlo
print(list(filter(es_par, numeros)))
```

### $map()$

Aplica una funcin a cada elemento de la lista. Devuelve una nueva lista con los resultados, por tanto, debemos asignar los resultados a una lista.

```python
# funcion que eleva el cuadrado un numero
def cuadrado(n):
    return n*n
```

```python
# lista de numeros
numeros = [1, 7, 3, 22, 5, 11, 5, 8, 12, 0]

# aplicar map para elevar el cuadrado cada numero
numeros_cuadrados = list(map(cuadrado, numeros))
print(numeros_cuadrados)
```

### $zip()$

Agrupa elementos de varias listas y en tuplas contenidas en una lista

```python
# lista de nombres y edades
nombres = ["Ana", "Juan", "Maria", "Pedro"]
edades = [25, 30, 22, 35]
```

```python
# aplicamos zip para combinar las dos listas
combinar = list(zip(nombres, edades))
print(combinar)
```

Si las listas son de diferentes tamaos, zip solo combina hasta el menor tamano de las listas, ignorando los elementos restantes de la lista ms larga

```python
nombres = ["Ana", "Juan", "Maria"]
edades = [25, 30]
```

```python
combinar = list(zip(nombres, edades))
print(combinar)
```
