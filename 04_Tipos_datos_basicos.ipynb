# **Tipos de Datos Bsicos**

El tipo representa el tipo de valor y determina cmo se puede utilizar el valor. Todo en Python es un objeto y cada objeto tiene una identidad, un tipo y un valor.

Para determinar el tipo de una variable, utilizamos la funcin _type()_

```python
a = 'string'
print(type(a))
```

El valor de algunos objetos se pueden cambiar. Los objetos cuyo valor se puede cambiar se denominan "_mutables_" y los objetos cuyo valor no cambian luego de ser creado se denominan "_inmutables_".

# __Enteros__

Son un tipo de datos que permite representar nmeros enteros, es decir, positivos y negativos no decimales.

```python
num = 12
print(num)
```

En Python, a partir de la versin 3, se pueden representar nmeros enteros tanto grande como pequeos ya que por debajo asigna la memoria necesaria al nmero que se va a representar.

```python
num = 220**250
print(num)
print(type(num))
```

### _Enteros de diferentes base_

Tambin es posible asignar valores en binario, hexadecimal y octal. 

- Prefijo 0b para representar un nmero binario.
- Prefijo 0x para representar un nmero hexadecimal.
- Prefijo 0o pra representar un nmero octal.

Al imprimir tales nmeros, se convierten en enteros para todos los casos.

```python
num_b = 0b101
num_h = 0x17
num_c = 0o720
```

```python
print(num_b, type(num_b))
print(num_h, type(num_h))
print(num_c, type(num_c))
```

### _Flexibilidad en la memoria_

Python asigna diferentes nmeros de bits a las variables en funcin del nmero que quieren representar. Vamos a utilizar la funcin _getsizeof()_ para mostrar el tamao de una variable en memoria.

```python
# libreria para usar la funcion
import sys

# definimos dos numeros
num1 = 5**10000
num2 = 10
```

```python
# mostramos el tamano en memoria de cada numero
print(sys.getsizeof(num1), type(num1))
print(sys.getsizeof(num2), type(num2))
```

Hay casos limites donde el nmero a representar es tan grande que genera una excepcin.

```python
print(5e200**2)
```

Un caso inusual es que si intentamos representar un nmero mayor que el anterior nos devuelve lo siguiente en vez de una excepcin. Representa el infinito.

```python
print(2e2000**2)
```

### _Convertir a int_

Permite convertir otro tipo de dato a entero. Como _int_ no puede contener decimales, al intentar convertir un nmero decimal se truncar todo lo que est a la derecha de la coma.

```python
num = int(2.67)
print(num, type(num))
```

# __Float__

Permite representar un nmero positivo o negativo con decimales (nmeros reales). Al declarar un variable y asignarle un valor decimal, por defecto la variable ser de tipo float.

```python
num = 0.0032
print(num, type(num))
```

Tambin se pueden declarar usando la notacin cientfica con '_e_' y el exponente. Vamos a representar a 1.93 por diez elevado a la -3.

```python
num = 1.93e-3
print(num, type(num))
```

### _Conversin a float_

Podemos convertir otro tipo a float mediante la funcin _float()_.

```python
a = float(True)
b = float(2)

print(a, type(a))
print(b, type(b))
```

En este caso, _True_ es en realidad tratado como 1.0 al convertirlo a float.

### _Rango Representable_

Los floats no tienen precisin infinita. Veamos el siguiente ejemplo:

```python
num = 0.9999999999999999999
print(num, type(num))
print(1 == num)
```

Los float tienen unos valores mnimo y mximo que pueden representar:

- Precisin mnima 2.2250738585072014e-308.
- Precisin mxima 1.7976931348623157e+308.

Verifiquemoslo!

```python
import sys

print(sys.float_info.min)
print(sys.float_info.max)
```

Si asignamos un valor mayor que el _max_, la variable toma el valor _inf_ que representa infinito.

```python
num = 1.7976931348623157e+309
print(num)
```

### _Precisin del float_

Los computadores no pueden representar cualquier nmero y mucho menos si este es un irracional.

Al dividir _1/3_ debera resultar en 0.3 peridico, pero para Python es imposible representar tal nmero.

```python
print('{:10f}'.format(1.0/3.0))
```

La siguiente operacin debera ser cero, pero esto no es asi.

```python
print(0.1 + 0.1 + 0.1 - 0.3)
```

Las cadenas son conjuntos de caracteres inmutables, es decir, una vez que se crea no se puede cambiar ningn carcter dentro de ella.

Se utiliza para comprobar si un carcter o una subcadena est presente en una cadena o no. La expresin devuelve un booleano: True si concide y False de lo contrario.

### _Segmentacin de cadenas_

Consiste en cortar una subcadena de una cadena. Se hace a travs de los indices que corresponde a cada carcter de la cadena. 

- Se toman dos indices entre corchetes, separados por dos puntos _(:)_

- El primero indica la posicin desde donde se extrae la subcadena y el segundo donde finaliza. 

- Hay que considerar que el ltimo ndice no est incluido en el corte. 

```python
text = 'Librerias de Python'

# extraer la palabra 'Librerias'
word = text[0:9]

# mostrar palabra
print(word)
```

Tambin podemos acceder a travs de indices negativos.

```python
text = 'Librerias de Python'

# extraer la palabra 'Librerias' con indices negativos
word2 = text[-19:-10]

# mostrar palabra
print(word2)
```

### _Formateo de cadenas_

__Primera forma__

La primera forma de declarar una cadena que contenga variables en su interior es usando el operador _+_. Es importante recordar que la funcin _str()_ convierte en string lo que se pasa como parmetro.

```python
x = 10
edad = "Tu edad es: " + str(10)
print(edad)
```

__Segunda forma__

La segunda forma es usando el operador de cadenas _%_. Funciona con el uso de _%_ a la izquierda indicando el tipo que se quiere imprimir: _%s_ para string, _%d_ para enteros y _%f_ para decimales (Para mayor informacin sobre este tipo de formateo haz clic [aqui](https://docs.python.org/3/library/stdtypes.html#printf-style-string-formatting)) y a la derecha el _%_ la variable a imprimir.

```python
# declaramos la variable x
x = 5

# declaramos la variable a imprimir
s = 'El numero es: %d' %x

print(s)
```

Si tenemos ms de una variable, tambin se puede hacer pasando los parmetros dentro de _()_. 

```python
# variable a imprimir
s = 'Los numeros son: %d y %d' %(2, 8)
print(s)
```

Una forma moderna de realizar lo mismo, es usando _format()_

```python
# variable a imprimir
s = 'Hola {}. Tu edad es {}'.format('Miguel', 19)
print(s)
```

Es posible darle nombre a cada elemento y _format()_ se encargara de reemplazar todo

```python
s = 'Los numeros son {a} y {b}'.format(b=23, a=1)
print(s)
```

__Tercera forma__

Es una forma incluida en la version 3.6 de Python y recibe el nombre de cadenas literales o _f-string_. Permite incrustar expresiones dentro de candenas.

```python
# declaramos dos variables
a = 2; b = 3

# declaramos variable a imprimir
s = f'Los numeros son {a} y {b}'
print(s)
```

Incluso admite operaciones de las variables dentro de los corchetes

```python
# declaramos dos variables
a = 5; b = 7

# declaramos variable a imprimir
s = f'a + b =  {a+b}'
print(s)
```

Hasta se puede llamar una funcin dentro de los corchetes

```python
# declaramos una funcion
def funcion():
    return 12

# variable para mostrar el resultado
s = f'Al sumar 4 a la funcion nos da: {funcion() + 4}'
print(s)
```

### _Mtodos de String_

Se aplica sobre una cadena y la devuelve con su primera letra en mayscula.

```python
text = 'nueva oracion'
print(text.capitalize())
```

- __lower()__

Convierte todos los caracteres alfabticos en minscula.

- __swapcase()__

Convierte los caracteres alfabticos con mayscula en minscula y vicerversa.

- __upper()__

Convierte todos los caracteres alfabticos en mayscula.

```python
text = 'Nueva Oracion'
print(text.upper())
```

- __count()__

Permite contar las veces que otra cadena se encuentra dentro de la primera. Admite dos parmetros opcionales que indican donde empezar (_start_) y terminar (_end_) de buscar.

    count(sub[start, end])

```python
text = 'locura y tortura'
print(text.count('ura'))
```

```python
# buscar 'ura' hasta el septimo caracter
print(text.count('ura',1,7))
```

Devuelve True si la cadena esta formada nicamente por caracteres alfanumrico, False de lo contrario.Caracteres como @ y & no son alfanumricos.

```python
text = 'correo@dominio.com'
print(text.isalnum())
```

- __isalpha()__

Devuelve true si todos los caracteres son alfabticos, False de lo contrario. El espacio no cuenta como carcter alfabtico.

```python
text = 'Laluciernaga'
print(text.isalpha())
```

- __strip()__

Elimina a la izquierda y derecha el carcter que se le introduce. Si se llama sin parmetros elimina los espacios. Muy til para limpiar cadenas.

```python
text = '   aaa   '
print(text.strip())
```

- __zfill()__

Rellena la cadena con ceros a la izquierda hasta llegar a la longitud pasada como parmetro.

```python
text = '235'
print(text.zfill(7))
```

- __join()__

Devuelve la primera cadena unida a cada uno de los elementos de la lista que se le pasa como parmetro.

```python
text = ' y '.join(['vaca', 'perro', 'gato'])
print(text)
```

- __split()__

Divide una cadena en subcadenas y las devuelve almacenadas en una lista. La divisin es realizada de acuerdo al primer parmetro y el segundo parmetro indica el nmero mximo de divisiones a realizar.

```python
text = 'uno,dos,tres'
print(text.split(','))
```

# __Nmeros Complejos__

Son aquellos que tienen dos partes:

- Una parte real como $3$ o $1.9$.

- Una parte imaginaria como $3j$ o $7j$

La parte imaginaria viene acompaada de $j$ aunque tambin es comn usar la $i$. En Python, los nmeros complejos pueden ser creados sin tener que importar ninguna librera tal como sigue a continuacin.

```python
# declaramos el numero complejo
c = 3 + 8j

# mostramos en pantalla
print(c)

# mostramos su tipo
print(type(c))
```

Tambin se pueden crear haciendo uso de _complex_ pero sin usar la $j$

```python
c = complex(3,8)
print(c)
```

### _Accediendo a las partes_

Vamos a acceder a la parte real de un nmero imaginario.

```python
# declaramos el numero complejo
c = 27 + 3j

# acceder a la parte real
print(c.real)
```

Ahora, accedemos a la parte imaginaria

```python
# acceder a la parte imaginaria
print(c.imag)
```

### _Operaciones con nmeros complejos_

__Suma de complejos__

Se suman las partes reales por un lado y las imaginarias por otro.

```python
a = 21 + 3j
b = 12 + 9j

# mostrar suma
print(a+b)
```

__Resta de complejos__

Se restan las partes reales por un lado y las imaginarias por otro

```python
a = 21 + 3j
b = 12 + 9j

# mostrar resta
print(a-b)
```

__Multiplicacin de complejos__

Si mutiplicamos $a+bj$ por $c+dj$ el resultado para la parte real es $ac - bd$ y para la imaginaria es $ad + bc$

```python
a = 21 + 3j
b = 12 + 9j

# mostrar multiplicacion
print(a*b)
```

__Divisin de complejos__

Python tambin admite la divisin de dos nmeros complejos.

```python
a = 21 + 3j
b = 12 + 9j

# mostrar division
print(a/b)
```

__Conjugado de complejos__

Calcular el conjugado consiste en negar la parte imaginaria, es decir cambiar el signo de $+$ a $-$ y viceversa. Esto se realiza en Python con el mtodo _conjugate()_

```python
a = 21 + 3j

# mostrar el conjugado
print(a.conjugate())
```

> Para realizar mas operaciones con nmeros complejos puedes utilizar la librera [_cmath_](https://docs.python.org/3/library/cmath.html)
