# **Expresiones en Python**

## Clasificación


Se clasifican según el tipo de dato y los operadores que utilizan.

- $\text{Aritméticas}$: Análogas a fórmulas matemáticas, en donde se utilizan operadores aritméticos ($\ +\ ,\ -\ ,\ *\ ,\ /\ ,\ **\ ,\ \%\ ,\ //\ $), variables y constantes numéricas. Ejemplo: 

<br><br>

- $\text{Expresión algebraica}$:

$$-8 * 2 + \frac{4}{2}* \pi *\sqrt(x)$$
<br><br>

- $\text{Alfanuméricas}$: Se utilizan operadores alfanuméricos como la **concatenación** y producen resultados de tipo alfanuméricos. Ejemplo: "Ing.\ " + "Jorge\ López"\ =\ "Ing.\ Jorge\ López"

<br><br>

- $\text{Relacionales}$: Se utilizan operadores relacionales ($<\ ,\ >\ ,\ <=\ ,\ >=\ ,\ ==\ ,\ !=$). Con estos operadores se compara entre dos valores del mismo tipo, es decir que los operandos son ambos numéricos, ambos caracter o ambos de tipo lógico. El resultado de la evaluación de una expresión relacional es de tipo lógico, $True$ o $False$. Ejemplo:   

$$A + B < = A * B$$

<br><br>

- $\text{Lógicas}$: Se utilizan operadores lógicos ($and\ ,\ or\ y\ not$). El resultado de este tipo de expresión es un valor lógico, $True$ o $False$, el cual se deduce de analizar la tabla de verdad correspondiente a los conectores lógicos.

## Precedencia de operadores


Hace referencia a la prioridad entre operadores. Si no tengo signos de agrupación, ¿Qué opero primero?

La siguiente piramide muestra la prioridad que sigue Python de acuerdo a los operadores que contenga la expresión. Va de mayor prioridad en la parte mas alta de la piramide hasta los de menor prioridad en la base.

precedencia_operadores2.png

```python
(10/5) ** 4 > 15 and (8 + 4) == 12
# 2   ** 4 > 15 and 12       == 12
#     16   > 15 and     12   == 12
#       True    and    True
#              True
```

### Ejercicio 6

Encuentra la pendiente y la distancia euclideana entre los puntos ($2,2$) y ($6,10$)

```python
# pendiente = (y2 - y1)/(x2 - x1)
slope = (10 - 2)/(6 - 2)

# distancia euclidiana = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
distance = ((6 - 2)**2 + (10 - 2)**2)**0.5

print(f'Pendiente: {slope}, Distancia: {distance}')
```
